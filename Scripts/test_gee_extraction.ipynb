```python
ee.Initialize()
# Feature class that holds basins (this one is the hydrobasin 6 that connect to P4W reservoirs)
basins = ee.FeatureCollection(r'users/samanthakuzma/ReservoirWatch/P4W_hy6')

# Set the date range to run Starting date
original_start_date = datetime.date(2018, 1, 1) # Set start date
py_start_date = original_start_date # Create a copy that will update through each run
py_end_date = datetime.date(2020, 12, 31) # Set an end date to loop until
delta = datetime.timedelta(days=1) # Set date increment (i.e. 1 day increments)

while py_start_date <= py_end_date:
#     print(py_start_date)
    # Turn start date into GEE date
    startDate = ee.Date.fromYMD(py_start_date.year, py_start_date.month, py_start_date.day)
    endDate = startDate.advance(1, 'day')
    # Read in CHIRPS data (filter by date)
    chirps = ee.ImageCollection("UCSB-CHG/CHIRPS/DAILY").filter(ee.Filter.date(startDate, endDate))
    # Turn image collection into Image
    listOfImages = chirps.toList(chirps.size());
    firstImage = ee.Image(listOfImages.get(0));
    # Run zonal statistics
    geemap.zonal_statistics(firstImage, basins, "daily_chirp.csv", statistics_type='MEAN', scale = 5000)
    # Read in results to dataframe
    stats = pd.read_csv("daily_chirp.csv", header = 0, index_col = 4)
    stats = stats.filter(['mean'])
    stats['date'] = py_start_date
    # Either create a new dataframe at the start of the loop or...
    if original_start_date == py_start_date:
        df_final = stats.copy()
    #...append onto dataframe
    else:
        df_final = df_final.append(stats)
    # Move date up by increment
    py_start_date += delta

df_play = df_final.copy()
df_play.reset_index(inplace = True)
df_play.sort_values(['PFAF_ID', 'date'], inplace=True)
df_play.to_csv('daily_chirps_201801-202012.csv')
```
